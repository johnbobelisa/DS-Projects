<h1>Data Preprocessing

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Convert categorical variable to numerical for regression
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'No': 0, 'Yes': 1})
df['Extracurricular Activities']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Split data into features and target
features = df.drop(columns = 'Performance Index')
target = df['Performance Index']

features.head()
```

<h1>Exploratory Data Analysis</h1>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Plot histograms and Q-Q plots
for feature in features:
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Histogram of {feature}')
    
    plt.subplot(1, 2, 2)
    stats.probplot(df[feature], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot of {feature}')
    
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

<h1>Modeling</h1>

```python
from sklearn.model_selection import train_test_split

# Define features and target
X = features
y = target

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)
```

```python
from sklearn.linear_model import LinearRegression

# Instantiate the model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, root_mean_squared_error

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'R2: {r2}')
```

```python
model.score(X_train, y_train)
```

```python
import numpy as np
predict = np.round(model.predict(X_test), decimals = 1)
pd.DataFrame({"Actual Performance" : y_test, "Predicted Performance" : predict})
```

***Final equation is:***

2.85  Hours Studied + 1.02  Previous Scores + 0.60  Extracurricular Activities + 0.49  Sleep Hours + 0.19  Sample Question Papers Practiced - 34.06

```python
def predict_performance(hours_studied, previous_scores, extracurricular_activities, sleep_hours, sample_question_papers_practiced):
    # Coefficients and intercept
    intercept = -34.06
    coef_hours_studied = 2.85
    coef_previous_scores = 1.02
    coef_extracurricular_activities = 0.60
    coef_sleep_hours = 0.49
    coef_sample_question_papers_practiced = 0.19

    # Predict the performance index
    performance_index = (
        intercept +
        (coef_hours_studied * hours_studied) +
        (coef_previous_scores * previous_scores) +
        (coef_extracurricular_activities * extracurricular_activities) +
        (coef_sleep_hours * sleep_hours) +
        (coef_sample_question_papers_practiced * sample_question_papers_practiced)
    )

    return performance_index
```

```python
# Take user inputs
hours_studied = float(input("Enter hours studied: "))
previous_scores = float(input("Enter previous scores: "))
extracurricular_activities = float(input("Enter extracurricular activities (0 for No, 1 for Yes): "))
sleep_hours = float(input("Enter sleep hours: "))
sample_question_papers_practiced = float(input("Enter number of sample question papers practiced: "))

# Predict performance
performance_index = predict_performance(
    hours_studied,
    previous_scores,
    extracurricular_activities,
    sleep_hours,
    sample_question_papers_practiced
)

# Display the result
print(f"\nPredicted Performance Index: {performance_index:.2f}")
```
