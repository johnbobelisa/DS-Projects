***Data Preprocessing***

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory

DATA_DIR = r'C:\Users\Asus\OneDrive\Documents\Brain Tumor Detection Model\brain_tumor_dataset'
IMG_SIZE = (250, 250)
BATCH_SIZE = 32

# Load training dataset with binary labels
train_dataset = image_dataset_from_directory(
    DATA_DIR,
    labels='inferred',  # Automatically infer labels from directory structure
    label_mode='binary',  # Binary labels for binary classification
    validation_split=0.2,  # Use 20% of the data for validation
    subset="training",  # Load training subset
    seed=69,  # Seed for reproducibility
    image_size=IMG_SIZE,  # Resize images to 250x250
    batch_size=BATCH_SIZE,  # Batch size of 32
    color_mode='grayscale',  # Load images as grayscale
    shuffle=True,  # Shuffle the data, prevent overfitting
    interpolation='bilinear',  # Bilinear interpolation for resizing
    follow_links=False,  # Do not follow symlinks
    verbose=True  # Display data loading information
)

# Load validation dataset with binary labels
validation_dataset = image_dataset_from_directory(
    DATA_DIR,
    labels='inferred',  # Automatically infer labels from directory structure
    label_mode='binary',  # Binary labels for binary classification
    validation_split=0.2,  # Use 20% of the data for validation
    subset="validation",  # Load validation subset
    seed=69,  # Seed for reproducibility
    image_size=IMG_SIZE,  # Resize images to 250x250
    batch_size=BATCH_SIZE,  # Batch size of 32
    color_mode='grayscale',  # Load images as grayscale
    shuffle=True,  # Shuffle the data, prevent overfitting
    interpolation='bilinear',  # Bilinear interpolation for resizing
    follow_links=False,  # Do not follow symlinks
    verbose=True  # Display data loading information
)

# Optional: Prefetching for performance optimization
AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.prefetch(buffer_size=AUTOTUNE)

# Optional: Data augmentation for training
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1)
])

# Apply data augmentation to the training dataset
train_dataset = train_dataset.map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE)
```

***Creating the model***

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization

# Define the custom CNN model
model = Sequential([
    
    Input(shape=(250, 250, 1)),
    
    # First Convolutional Block
    Conv2D(32, (3, 3), activation='relu'),
    BatchNormalization(),  # Normalizes the output to have mean 0 and std 1
    MaxPooling2D(pool_size=(2, 2)),  # Reduces spatial dimensions from 250x250 to 125x125
    Dropout(0.25),

    # Second Convolutional Block
    Conv2D(64, (3, 3), activation='relu'),  # 64 filters to detect more complex features
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),  # Reduces spatial dimensions further from 125x125 to 62x62
    Dropout(0.25),

    # Third Convolutional Block
    Conv2D(128, (3, 3), activation='relu'),  # 128 filters for even more complex features
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),  # Reduces spatial dimensions further from 62x62 to 31x31
    Dropout(0.25),

    # Flattening and Dense Layers
    Flatten(),  # Flattens the 3D output to 1D vector
    Dense(128, activation='relu'),  # Dense layer to aggregate features
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Output layer with sigmoid activation for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()
```

***Train the model***

```python
import os
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# Define the directory and file name for saving the model checkpoint
checkpoint_dir = r'C:\Users\Asus\OneDrive\Documents\Brain Tumor Detection Model\models'
checkpoint_file = os.path.join(checkpoint_dir, 'brain_tumor_detector_model_v1.keras')

# Create the directory if it doesn't exist
os.makedirs(checkpoint_dir, exist_ok=True)

# Callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint(checkpoint_file, save_best_only=True, monitor='val_accuracy', mode='max')
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.00001)

# Training the model
history = model.fit(
    train_dataset,
    validation_data=validation_dataset,
    epochs=30,  # Starting with 30 epochs
    batch_size=BATCH_SIZE,  # Batch size of 32
    callbacks=[early_stopping, model_checkpoint, reduce_lr]
)
```

***Evaluate the model***

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, precision_recall_fscore_support
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Evaluate the model on the validation dataset
val_loss, val_accuracy = model.evaluate(validation_dataset)

# Predictions on the validation dataset
y_true = []
y_pred_probs = []

for images, labels in validation_dataset:
    preds = model.predict(images)
    y_true.extend(labels.numpy())
    y_pred_probs.extend(preds)

# Find the optimal threshold
def find_optimal_threshold(y_true, y_pred_probs):
    thresholds = np.arange(0.0, 1.0, 0.01)
    f1_scores = []
    precision_scores = []
    recall_scores = []
    
    for threshold in thresholds:
        y_pred_binary = (np.array(y_pred_probs) >= threshold).astype(int)
        precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred_binary, average='binary')
        precision_scores.append(precision)
        recall_scores.append(recall)
        f1_scores.append(f1)
    
    # Find the threshold that gives the maximum F1 score
    optimal_idx = np.argmax(f1_scores)
    optimal_threshold = thresholds[optimal_idx]
    
    return optimal_threshold, thresholds, precision_scores, recall_scores, f1_scores

# Convert predictions to binary labels with a default threshold of 0.5
y_pred_binary_default = (np.array(y_pred_probs) >= 0.5).astype(int)

# Find optimal threshold
optimal_threshold, thresholds, precision_scores, recall_scores, f1_scores = find_optimal_threshold(y_true, y_pred_probs)

# Convert predictions to binary labels with the optimal threshold
y_pred_binary_optimal = (np.array(y_pred_probs) >= optimal_threshold).astype(int)

# Plot the metrics against thresholds
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precision_scores, label='Precision')
plt.plot(thresholds, recall_scores, label='Recall')
plt.plot(thresholds, f1_scores, label='F1 Score')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.legend()
plt.title('Precision, Recall, and F1 Score vs. Threshold')
plt.show()

print(f'Optimal Threshold: {optimal_threshold}')

# Confusion Matrix for the default threshold
cm_default = confusion_matrix(y_true, y_pred_binary_default)
sns.heatmap(cm_default, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix (Threshold = 0.5)')
plt.show()

# Classification Report for the default threshold
print('Classification Report (Threshold = 0.5):')
print(classification_report(y_true, y_pred_binary_default))

# Confusion Matrix for the optimal threshold
cm_optimal = confusion_matrix(y_true, y_pred_binary_optimal)
sns.heatmap(cm_optimal, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title(f'Confusion Matrix (Optimal Threshold = {optimal_threshold})')
plt.show()

# Classification Report for the optimal threshold
print(f'Classification Report (Optimal Threshold = {optimal_threshold}):')
print(classification_report(y_true, y_pred_binary_optimal))

# ROC Curve and AUC
fpr, tpr, _ = roc_curve(y_true, y_pred_probs)
auc = roc_auc_score(y_true, y_pred_probs)
plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()

# Training and Validation Curves
history_dict = history.history
acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(acc) + 1)

plt.figure()
plt.plot(epochs, acc, 'bo', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

***Testing the model (random image selection)***

```python
import os
import random
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import numpy as np

# Optimal threshold
optimal_th = 0.56

# Define the directory
data_dir = r'C:\Users\Asus\OneDrive\Documents\Brain Tumor Detection Model\brain_tumor_dataset'
categories = ['no', 'yes']

# Function to load and preprocess image
def load_and_preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(250, 250), color_mode='grayscale')
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    return img_array

# Function to make a prediction and display the image with annotations
def display_predictions(model, data_dir, categories, num_images=1, images_per_row=2, optimal_threshold=0.56):
    num_rows = (num_images + images_per_row - 1) // images_per_row  # Calculate number of rows needed
    fig, axes = plt.subplots(num_rows, images_per_row, figsize=(15, 5 * num_rows))
    axes = axes.flatten()  # Flatten the axes array for easy iteration

    previous_category_index = 1  # Start with the "yes" category
    for i in range(num_images):
        # Alternate category selection
        current_category_index = 1 - previous_category_index
        category = categories[current_category_index]
        previous_category_index = current_category_index
        
        category_path = os.path.join(data_dir, category)
        img_name = random.choice(os.listdir(category_path))
        img_path = os.path.join(category_path, img_name)

        # Load and preprocess the image
        img_array = load_and_preprocess_image(img_path)
        
        # Make prediction
        prediction = model.predict(img_array)[0][0]
        
        # Debugging: Print prediction probability
        print(f'Prediction probability for {img_name}: {prediction}')
        
        predicted_label = 'yes' if prediction >= optimal_threshold else 'no'
        
        # Actual label
        actual_label = category

        # Load image for display
        img = image.load_img(img_path, target_size=(250, 250), color_mode='grayscale')

        # Plot the image
        ax = axes[i]
        ax.imshow(img, cmap='gray')
        ax.axis('off')
        
        # Display file name
        ax.text(10, -10, f'File: {img_name}', fontsize=12, color='blue', fontweight='bold')
        
        # Display actual label at the top
        ax.text(10, 10, f'Actual: {actual_label}', fontsize=12, color='white', fontweight='bold', bbox=dict(facecolor='blue', alpha=0.5))
        
        # Display model prediction at the bottom
        ax.text(10, 230, f'Prediction: {predicted_label}', fontsize=12, color='white', fontweight='bold', bbox=dict(facecolor='blue', alpha=0.5))

        # Compare and display result
        result_text = 'AI is correct!' if actual_label == predicted_label else 'AI is incorrect!'
        result_color = 'green' if actual_label == predicted_label else 'red'
        ax.text(10, 260, result_text, fontsize=14, color=result_color, fontweight='bold')

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# Assuming the model is already loaded as `model`
display_predictions(model, data_dir, categories, num_images=4, images_per_row=2, optimal_threshold=optimal_th)
```
