***Splitting the dataset***

```python
X = df.drop('Class', axis=1) # Features
y = df['Class'] # Target

X
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

***Evaluate the model***

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classf_report = classification_report(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression

# Load the dataset
df = pd.read_csv('breast_cancer.csv')

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# List of features to plot
features = [
    'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 
    'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 
    'Bland Chromatin', 'Normal Nucleoli', 'Mitoses'
]

# Plot settings
fig, axes = plt.subplots(3, 3, figsize=(20, 15))
axes = axes.flatten()

for i, feature in enumerate(features):
    ax = axes[i]
    
    # Scatter plot of feature vs. Tumor Type
    sns.scatterplot(x=feature, y='Class', data=df, alpha=0.6, edgecolor='k', ax=ax)
    
    # Generate a range of feature values
    x_values = np.linspace(df[feature].min(), df[feature].max(), 300)
    
    # Fit logistic regression model for the feature
    model = LogisticRegression(max_iter=200)
    model.fit(df[[feature]], df['Class'])
    
    # Compute probabilities using the logistic regression model
    linear_combination = model.intercept_ + model.coef_[0] * x_values
    probabilities = sigmoid(linear_combination)
    
    # Plot the logistic regression curve
    ax.plot(x_values, probabilities, color='#557153', linewidth=1.2)
    
    # Customize the plot
    ax.set_title(f"Breast Cancer Diagnosis \nusing {feature}", fontsize=15, fontweight='bold', pad=20)
    ax.set_xlabel(feature, fontsize=12, labelpad=10)
    ax.set_ylabel("Tumour type", fontsize=12, labelpad=10)
    ax.set_yticks([0, 1])
    ax.set_yticklabels(["Benign", "Malignant"], fontsize=10)
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=10)
    ax.set_ylim(0, 1)
    ax.set_facecolor('#F0EBCE')

    # Remove grid lines
    ax.grid(False)

# Remove empty subplot(s)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.gcf().set_facecolor('#F0EBCE')
plt.show()
```
